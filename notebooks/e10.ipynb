乱数A : 標準偏差と分散が1となるのは、標準正規分布とポワソン分布であり、ポワソン分布は min, max = 0, 1 ので $X \sim N(0, \ 1)$

```python
n = 1000
np.random.seed()
df = pd.DataFrame()
df = pd.concat([df, pd.DataFrame(stats.bernoulli.rvs(0.8, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.uniform.rvs(0, 10, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.poisson.rvs(1, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.norm.rvs(size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.binom.rvs(15, 0.5, size=n)).describe().T], axis=0)
df.index = ['bernoulli', 'uniform', 'poisson', 'norm', 'binom']
df
```

乱数B : 最小値 0, 中央値 3, より $\lambda = 3$ のポワソン分布

```python
df = pd.DataFrame()
df = pd.concat([df, pd.DataFrame(stats.bernoulli.rvs(0.8, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.uniform.rvs(0, 15, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.poisson.rvs(3, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.norm.rvs(3, 5, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.binom.rvs(10, 0.5, size=n)).describe().T], axis=0)
df.index = ['bernoulli', 'uniform', 'poisson', 'norm', 'binom']
df
```

乱数C : 最小値 ≒ 0, 最大値 ≒ 3, より $X \sim U(0, 3)$ の一様分布

```python
df = pd.DataFrame()
df = pd.concat([df, pd.DataFrame(stats.bernoulli.rvs(0.3, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.uniform.rvs(0, 3, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.poisson.rvs(3, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.norm.rvs(4, 10, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.binom.rvs(5, 0.5, size=n)).describe().T], axis=0)
df.index = ['bernoulli', 'uniform', 'poisson', 'norm', 'binom']
df
```

乱数D : min, max, range = 0, 1, 1 より $X \sim Be(0.5)$

```python
df = pd.DataFrame()
df = pd.concat([df, pd.DataFrame(stats.bernoulli.rvs(0.5, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.uniform.rvs(0, 10, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.poisson.rvs(2, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.norm.rvs(0, 1, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.binom.rvs(12, 0.5, size=n)).describe().T], axis=0)
df.index = ['bernoulli', 'uniform', 'poisson', 'norm', 'binom']
df
```

乱数E : min, max = 1, 5 より $X \sim b(5, 0.8)$

```python
df = pd.DataFrame()
df = pd.concat([df, pd.DataFrame(stats.bernoulli.rvs(0.85, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.uniform.rvs(1, 3, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.poisson.rvs(1, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.norm.rvs(0, 1, size=n)).describe().T], axis=0)
df = pd.concat([df, pd.DataFrame(stats.binom.rvs(5, 0.8, size=n)).describe().T], axis=0)
df.index = ['bernoulli', 'uniform', 'poisson', 'norm', 'binom']
df
```
