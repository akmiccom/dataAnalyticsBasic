```python
from glob import glob
import pandas as pd
```

```python
files = glob("../csv/例題_データアナリティクス基礎/*")
# for i, f in enumerate(files):
#     print(i, f)
```

```python
purchase_data = pd.read_excel(files[3])
purchase_data.columns = ['id', 'customer_id', 'price', 'food', 'weekday', 'gender']
purchase_data.head()
```

パレート表

```python
item_count = purchase_data.pivot_table(index='food', values='customer_id', aggfunc='count')
item_count.rename(columns={'customer_id': 'count'}, inplace=True)
item_count = item_count.sort_values(by='count', ascending=False)
item_count['cumsum'] = item_count.cumsum()
item_count['ratio'] = item_count['count'] / item_count['count'].sum()
item_count['c_ratio'] = item_count['cumsum'] / item_count['count'].sum()
item_count
```

```python
item_price = purchase_data.pivot_table(index='food', values='price', aggfunc='sum')
item_price = item_price.sort_values(by='price', ascending=False)
item_price['cumsum'] = item_price.cumsum()
item_price['ratio'] = item_price['price'] / item_price['price'].sum()
item_price['c_ratio'] = item_price['cumsum'] / item_price['price'].sum()
item_price
```
