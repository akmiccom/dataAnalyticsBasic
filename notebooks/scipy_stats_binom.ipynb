- scipy を使った二項分布

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import japanize_matplotlib
```

二項分布のモーメント

```python
# Culculate the first moments.
n, p = 10, 0.5
mean, var, skew, kurt = stats.binom.stats(n, p, moments='mvsk')
print(f'mean: {mean}', f'var: {var}', f'skew: {skew:.04f}', f'kurt: {kurt:.04f}')
```

n, p = 10, 0.5 の 二項分布

```python
n, p = 10, 0.5
attempt = np.arange(n+1)
dist = stats.binom.pmf(attempt, n, p)
print(dist)
plt.figure(figsize=(5, 3))
plt.bar(attempt, dist)
plt.xlabel('number of successes')
plt.ylabel('probability')
plt.title(f'n={n}, p={p} の二項分布')
plt.show()
```

いくつかの二項分布

```python
def binorm(n, p):
    attempt = np.arange(n + 1)
    dist = stats.binom.pmf(attempt, n, p)
    # print(dist)

    return attempt, dist


binom_data = [[10, 0.3], [10, 0.5], [10, 0.7]]
fig, axs = plt.subplots(1, len(binom_data), figsize=(len(binom_data) * 4, 3))
for (n, p), ax in zip(binom_data, axs):
    attempt, dist = binorm(n, p)
    ax.bar(attempt, dist)
    ax.set_xlabel("number of successes")
    ax.set_ylabel("probability")
    ax.set_title(f"n={n}, p={p} の二項分布")
plt.show()
```

```python
# Display the probability mass function(pmf)
plt.figure(figsize=(5, 3))
x = np.arange(stats.binom.ppf(0.01, n, p), stats.binom.ppf(0.99, n, p))
plt.plot(x, stats.binom.pmf(x, n, p), 'bo', ms=8, label='binom_pmf')
plt.vlines(x, 0, stats.binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
plt.legend()
plt.show()
```
